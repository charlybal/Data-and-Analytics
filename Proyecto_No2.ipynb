```python
from pandas.core.generic import Axis                                              #Para realizar la parte de agrgar nuevas columnas a mi DataFrame y realizar unos
from numpy.core.getlimits import log10                                            #cálculos tuvimos que importar 2 librerías "log10" y "Axis"
Am = df2['Millas_Por_Galon'].to_numpy()                                           #En esta parte separamos la variable 1 que es "Millas_por_Galon" y la variable 2 que es
Ac = df2['Cilindros'].to_numpy()                                                  # "Cilindros" de nuestro DataFrame y convertimos las columnas en "arrays".


V1 = log10(Am/Ac)                                                                 #realizamos operaciones para obtener mi primera columna nueva "V1"
V2 = np.sqrt(Am)*np.exp(Ac)/200                                                   #realizamos operaciones para obtener mi segunda columna nueva "V2"

V3 = list()                                                                       #Para mi tercera columna realizamos un ciclo anidado usando While e if "V3"
i = 0

while i < 406:
  if Am[i] > Ac[i]:
    V3.append(5)
  else:
    V3.append(3) 
  i += 1

hh = np.array(V3)
V3 = hh


V4 = (1/np.log2(Am/Ac))*100                                                       #realizamos operaciones para mi cuarta columna "V4"
V5 = (Ac**2)/Am                                                                   #Realizamos operaciones para obter la quinta columna "V5"

df2.insert(9, "V1", V1, False)                                                   #En esta parte agregamos las columnas arrays directamente a mi DaraFrame utiliando la
df2.insert(10, "V2", V2, False)                                                  # función "insert"
df2.insert(11, "V3", V3, False)
df2.insert(12, "V4", V4, False)
df2.insert(13, "V5", V5, False)

df2

```

```python
def analisism(x):                                                                 #En esta parte creamos una función para realizar un análisis de la "Media"
   m = np.mean(x)
   d = np.std(x)
   r = list()
   i = 0
   while i < 406:
     if x[i] > m+d:
       r.append('Alto')
     elif x[i] >= m-d: 
       r.append('Medio')
     else:
       r.append('Bajo')
     i += 1
   gg = np.array(r)
   r = gg
   return r   
```

```python
AV1 = analisism(V1)                                                               #Utilizamos la función creada para el análsis de la media y creamos
df2.insert(14, "Analisis_Media_V1", AV1, False)                                   #dos columnas nuevas
AV5 = analisism(V5)
df2.insert(15, "Analisis_Media_V5", AV5, False)
df2
```
